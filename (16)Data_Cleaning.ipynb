```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

ufo = pd.read_csv('https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/ufo.csv')
ufo.head()
```

```python
ufo.shape
```

```python
ufo.info()
```

```python
# check for missing values
ufo.isnull().sum()
```

```python
# dropna()
#remove missing values

ufo.dropna().shape
```

```python
# handel missing data
# fillna()
# fill NA/NaN values using specified method

ufo['City'].fillna(value='Citynotmentioned',inplace=True)
ufo['Colors Reported'].fillna(value='Colornotidentified',inplace=True)
ufo['Shape Reported'].fillna(value='shapenotidentified',inplace=True)
```

```python
ufo.isnull().sum()
```

```python
# filling missing data by taking up mean value
data = pd.read_csv('https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/sample11.csv')
data.head()
```

```python
mean_value = data['Percentage'].mean()
mean_value = round(mean_value,1)
data['Percentage'].fillna(value=mean_value,inplace=True)
data
```

```python
# Dealing With Duplicate Data
# Create a DataFrame
data = {
      'StudentName': ['Mark', 'Ali', 'Bob', 'John', 'Johny', 'Mark'],
      'Score': [45, 65, 76, 44, 39, 45]
}
df = pd.DataFrame(data)
df
```

```python
# check fr duplicate records
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
df[df['StudentName']=='John']
```

```python
# dropping duplicates

df.drop_duplicates(inplace=True)
df
```
