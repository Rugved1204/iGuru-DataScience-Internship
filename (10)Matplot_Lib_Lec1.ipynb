```python
import numpy as np
import pandas as pd
import matplotlib
print('Matplotlib Version: ;',matplotlib.__version__)
```

```python
import matplotlib.pyplot as plt
```

```python
# Line Graph
# plot(x,y) = draws a line from point to point
plt.plot([1,5,3],[8,2,9])
plt.show()
#show = used to display the graph
```

```python
x = [5,7,10]
y = [10,19,20]
plt.plot(x,y)
plt.title('Line Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# adding styles to graph
# https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
# style.available = used to show stylees available
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html


from matplotlib import style
style.use('ggplot')

x1 = [5,7,10]
x2 = [8,6,7]
y1 = [10,19,20]
y2 = [15,11,9]
plt.plot(x1,y1, color='green',linewidth=3, label='Product1', linestyle='dashed')
plt.plot(x2,y2, color='red',linewidth=3, label='Product2',linestyle='dotted')
plt.title('Line Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt. legend(loc=2) # help you to display the labels
plt.show()
```
