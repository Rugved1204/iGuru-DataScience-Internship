```python
# how to use pandas to deal with different-different file format like:
# Read an Excel-file
# Read an CSV-file
# Read an JSON-file
# Read an txt-file
# how to upload local file


import pandas as pd
```

```python
# Read an Excel-file
# read_excel() - reads data from sheet1 by default

data = pd.read_excel('https://raw.githubusercontent.com/yash240990/Python/master/sampleExcelData.xlsx')
data
```

```python
# how to know how many sheets are present in excel file?

excelfile = pd.ExcelFile('https://raw.githubusercontent.com/yash240990/Python/master/sampleExcelData.xlsx')
excelfile.sheet_names
```

```python
# To access sheet other than first sheet

data = pd.read_excel(excelfile, 'Sheet2')
data
```

```python
# Read a CSV file

data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Position_Salaries.csv')
data
```

```python
# Reading data in txt delimited format

data = pd.read_csv('https://www.inf.ed.ac.uk/teaching/courses/inf2b/labs/iris.txt', sep='\t')
data
```

```python
# How to read json format

data = pd.read_json('https://raw.githubusercontent.com/yash240990/Python/master/simple.json')
data
```
