```python
# Data Visualization - Seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print(sns.__version__)
```

```python
# built-in datasets in seaborn
sns.get_dataset_names()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
#line graph ---- lineplot()
sns.lineplot(x='sepal_length', y='sepal_width', data=iris)
plt.title('Line Plot in Seaborn')
plt.show()
```

```python
# Customize seaborn graph
sns.set_style('darkgrid' , {'grid.linestyle':'--' , 'grid.color':'k'} )    #line styles = -- , : , -
sns.lineplot(x='sepal_length', y='sepal_width', data=iris)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('sepal_length', fontsize=12)
plt.ylabel('sepal_width', fontsize=12)
plt.title('Line Plot in Seaborn')
plt.show()
```

```python
#change the figure size
plt.figure(figsize=(10,5))
sns.lineplot(x='sepal_length', y='sepal_width', data=iris)
plt.title('Line Plot in Seaborn')
```

```python
pokemon = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/pokemon.csv')
pokemon.head()
```

```python
#scatter plot
sns.scatterplot(x='Attack', y='Defense', data=pokemon , hue='Generation')
plt.title('Scatter Plot in Seaborn')
plt.figure(figsize=(15,5))
plt.show()
```

```python
pokemon.Generation.value_counts().sort_index()
```

```python
colors=['r','y','b','g','c','m']
sns.scatterplot(x='Attack',y='Defense',hue='Generation',palette=colors,data=pokemon)
plt.title('scatterplot in seaborn')
plt.figure(figsize=(15,5))
plt.show()
```
