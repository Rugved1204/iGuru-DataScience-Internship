```python
# Data Visualization - Seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
print(sns.__version__)
```

```python
# built-in datasets of seaborn
sns.get_dataset_names()
```

```python
dataset = sns.load_dataset('iris')
dataset
```

```python
# Line Graph --> lineplot()
sns.lineplot(x='sepal_length',y='sepal_width',data=dataset)
plt.title('Line Graph In Seaborn')
plt.show()
```

```python
# Customize Seaborn Plots
sns.set_style('whitegrid',{'grid.color':'k','grid.linestyle':':'})  # linestyle -- : -
sns.lineplot(x='sepal_length',y='sepal_width',data=dataset)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Line Graph In Seaborn')
plt.show()
```

```python
# Change the figure size
plt.figure(figsize=(10,5))
sns.lineplot(x='sepal_length',y='sepal_width',data=dataset)
plt.title('Line Graph In Seaborn')
plt.show()
```

```python
pokemon = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/pokemon.csv')
pokemon
```

```python
# Scatter Plot In Seaborn
sns.scatterplot(x='Attack',y='Defense',hue='Generation',data=pokemon)
plt.title('Scatter Plot In Seaborn')
plt.show()
```

```python
pokemon.Generation.value_counts().sort_index()
```

```python
colors = ['r','y','b','g','c','m']
sns.scatterplot(x='Attack',y='Defense',hue='Generation',palette=colors,data=pokemon)
plt.title('Scatter Plot In Seaborn')
plt.show()
```

```python
# lmplot()  -- used to generate a scatter plot using a regression line
colors = ['r','y','b','g','c','m']
sns.lmplot(x='Attack',y='Defense',hue='Generation',palette=colors,data=pokemon)
plt.title('Scatter Plot In Seaborn')
plt.show()
```

```python
# countplot() - generate a Bar Graph

plt.figure(figsize=(10,5))
sns.countplot(x='Type 1',data=pokemon)
plt.title('Count Plot In Seaborn')
plt.show()
```

```python
pokemon['Type 1'].value_counts()
```

```python
pkmn_type_colors = ['#3ae40c',  # Grass
                    '#e52106',  # Fire
                    '#0AE3D6',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#ECF709',  # Dragon
                    '#7A786C',  # Dark
                    '#D3D1C0',  # Steel
                    '#91E8EE',  # Flying
                   ]
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Type 1',hue='Type 1',palette=pkmn_type_colors,data=pokemon)
plt.title('Count Plot In Seaborn')
plt.xticks(rotation=45)
plt.show()
```

```python
# relplot()
# shows the relationship b/w two variables

tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill',y='tip',col='sex',hue='smoker',data=tips)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',col='time',hue='day',data=tips)
plt.show()
```
