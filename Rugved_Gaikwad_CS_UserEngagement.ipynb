```python
# DATA ANALYSIS USING DATA VISUALIZATION

# Responses based on age
age_counts = data["Age"].value_counts(normalize=True).sort_index() * 100
plt.figure(figsize=(10,5))
ax = sns.barplot(x=age_counts.index, y=age_counts.values, order=age_counts.index, palette="pastel")
ax.set(xlabel="Age", ylabel="Percentage (%)", title="Responses Based on Age", frame_on=False)
ax.tick_params(length=0)
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=12, color='black', fontweight='bold')
plt.show()
```

```python
# responses based on gender

colors=["skyblue", "lightcoral", "lightgreen"]
data.groupby('Gender').size().plot(kind='pie', autopct='%1.1f%%', colors=colors, startangle=90)
plt.title("Responses based on Gender")
plt.show()
```

```python
# user occupation based on gender

occupation_gender_counts = data.groupby(["Occupation", "Gender"]).size().reset_index(name="Count")
occupation_totals = occupation_gender_counts.groupby("Occupation")["Count"].sum()
occupation_gender_counts["Percentage"] = occupation_gender_counts.apply(
    lambda row: (row["Count"] / occupation_totals[row["Occupation"]]) * 100, axis=1
)
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=occupation_gender_counts, x="Occupation", y="Count", hue="Gender", palette="coolwarm")
for p, percentage in zip(ax.patches, occupation_gender_counts["Percentage"]):
    height = p.get_height()
    ax.annotate(f'{percentage:.1f}%', (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom', fontsize=11, color='black', fontweight='bold')
ax.set(xlabel="User Occupation", ylabel="Count", title="User Occupation Based on Gender")
```

```python
fig, axarr = plt.subplots(2, 2, figsize=(12, 8))
categories = [
    ("OnlineShoppingFrequency", "For Online Shopping"),
    ("ArticleReadingFrequency", "For Reading Articles"),
    ("VideoWatchingFrequency", "For Watching Videos"),
    ("OnlineDiscussionFrequency", "For Online Discussions")
]
for i, (category, title) in enumerate(categories):
    row, col = divmod(i, 2)
    counts = data[category].value_counts(normalize=True) * 100
    bars = counts.sort_index().plot.bar(ax=axarr[row][col], fontsize=8, color='royalblue')
    axarr[row][col].set_title(title, fontsize=15)
    for p in bars.patches:
        height = p.get_height()
        axarr[row][col].annotate(f'{height:.1f}%',
                                 (p.get_x() + p.get_width() / 2, height),
                                 ha='center', va='bottom', fontsize=10, fontweight='bold')
    axarr[row][col].set_ylabel("Percentage")
plt.subplots_adjust(hspace=0.6, wspace=0.4)
sns.despine()
plt.show()
```

```python
# Horizontal Bar Chart for Suggested Improvements
plt.figure(figsize=(10, 5))
improvement_counts = data["SuggestedImprovements"].value_counts(normalize=True).sort_values() * 100
bars = plt.barh(improvement_counts.index, improvement_counts.values, color=sns.color_palette("pastel"))
plt.title("Suggested Improvements", fontsize=14)
plt.xlabel("Percentage")
plt.grid(axis="x", linestyle="--", alpha=0.7)
for bar in bars:
    width = bar.get_width()
    plt.text(width + 1, bar.get_y() + bar.get_height()/2, f"{width:.1f}%", va="center", fontsize=10, fontweight="bold")
plt.show()
```

```python
#  Bar Chart for Requested Features
plt.figure(figsize=(10, 5))
features_counts = data["RequestedFeatures"].value_counts(normalize=True) * 100
bars = sns.barplot(x=features_counts.index, y=features_counts.values, palette="pastel")
plt.title("Requested Features", fontsize=17)
plt.ylabel("Percentage")
plt.xticks(rotation=30, ha='right')
plt.grid(axis="y", linestyle="--", alpha=0.7)
for bar, percentage in zip(bars.patches, features_counts.values):
    bars.annotate(f"{percentage:.1f}%",
                  (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                  ha='center', va='bottom', fontsize=10, fontweight="bold")
plt.show()
```

```python
# Pie Chart for Accessibility Rating
accessibility_counts = data["AccessibilityRating"].value_counts()
plt.pie(accessibility_counts, labels=accessibility_counts.index, autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Accessibility Rating Distribution", fontsize=14)
plt.show()
```

```python
# user rating

plt.figure(figsize=(8, 5))
rating_counts = data["UserExperienceRating"].value_counts(normalize=True) * 100
bars = sns.barplot(x=rating_counts.index, y=rating_counts.values, color="skyblue")
plt.title("User Experience Rating Distribution", fontsize=14)
plt.xlabel("Rating")
plt.ylabel("Percentage")
plt.grid(axis="y", linestyle="--", alpha=0.7)
for bar, percentage in zip(bars.patches, rating_counts.values):
    bars.annotate(f"{percentage:.1f}%",
                  (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                  ha='center', va='bottom', fontsize=10, fontweight="bold")
plt.show()
```
